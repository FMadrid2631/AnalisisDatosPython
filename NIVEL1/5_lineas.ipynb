# Grfico de Linea
Grfico que presenta valores numricos conectados por lineas de izquierda a derecha. Tiene un objetivo similar al grfico de barras. Nos sirve para mostrar diferencias de variables numricas a travs de mi variable en el Eje X. Sin embargo, es ms til cuando queremos mostrar tendencias de subida o de bajada de mi variable numrica. A diferencia del grafico de barras, el Eje X de izquierda a derecha suele representar una unidad de progresin ordenada de menor a mayor (ejemplo: das, aos, meses, horas; es decir, usualmente unidades de tiempo). Es decir, Variables categricas ordinales. A diferencia del grfico de barras, yo puedo tener muchas categoras o puntos en el Eje X ya que no necesito dibujar barras en cada punto. El grfico puede tener varias lneas que representan diferentes grupos. En este caso se denomina grfico multilinea. Y obviamente, cada lnea tiene que estar correctamente diferenciada.    
**Eje X**: Variable categrica ordinal     
**Eje Y**: Variable numrica     
**Otros nombres**: Cuando mi cantidad de valores en el Eje X es muy alta y es una variable temporal, se puede denominar Serie de Tiempo.



### Cuando utilizar
- Analizar tendencias de subida o de bajada en mis variables.
- Comparar diferencias de variables numricas entre uno o ms grupos en mis datos.
- Quiero visualizar cambios en magnitud a lo largo de mi Eje X. Sin embargo, si tengo pocos puntos, un grfico de barras es mejor alternativa si este es mi nico objetivo.


### Cuando no utilizar
- Si mi variable en el Eje X no tiene una progresin. En ese caso, tendra que pensar en un Grfico de Barras.
- Si tengo menos de 3 puntos para graficar.
- Si quiero comparar proporciones.
- Si quiero obtener informacin sobre distribuciones.


### Ejemplos de Uso   
- Si soy dueo de una pgina web, me gustara saber como ha cambiado el nmero de visitas entre diferentes aos
- En un banco, me interesara saber cuantas transacciones ocurren mensualmente


### Consejos
- No es necesario que el eje Y empiece en el valor del "cero" (a diferencia de los graficos de barras). He ahi la razn por la que no es tan til como el diagrama de barras para comparar magnitudes.
- No interpoles o suavices la curva. Esto podra dar una falsa ilusin de valores inexistentes. Siempre une cada punto por una lnea recta.
- Si no hay una cantidad grande de puntos, aade una clave visual para identificar correctamente a que parte de la lnea pertenece cada punto (un triangulo, un cuadrado, un circulo, una equis, etc).
- No utilices colores innecesariamente.
- Si vas a hacer un Grfico Multilinea para codificar una variable categrica adicional: 
  - Procura que no existan muchas lineas.
  - Procura que las variables que cada linea representan sean diferenciables (por ejemplo, a traves de colores diferentes, o a travs de una forma diferente en los puntos)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/tienda_ventas.csv', index_col='id')
df
```

```python
# Nueva columna con el formato: YYYY-MM
df['fecha'] = df['date'].apply(lambda x: x[:7])
df
```

```python
# Ventas por mes en todas las tiendas tomando en cuenta todos los productos
por_mes = df[['fecha', 'sales']].groupby('fecha').sum()
por_mes
```

```python
# otra opcion de Ventas por mes en todas las tiendas tomando en cuenta todos los productos
por_mes2 = df.groupby('fecha').agg(
    {
        'sales' : 'sum'
    }
)
por_mes2
```

```python
# Grafico de Lineas con Pandas
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
por_mes.plot.line(
    y='sales', 
    style='.-',
    linestyle='dotted', # cambia la linea por puntitos pequeos
    figsize=(10, 5)
)
plt.title('Ventas por mes por ao')
```

```python
# Grafico de Lineas con Pandas
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
por_mes2.plot.line(
    y='sales', 
    style='.-',
    figsize=(10, 5)
)
plt.title('Ventas por mes por ao')
```

```python
# Grafico de Barras con los mismos datos
por_mes.plot.bar(
    y='sales', 
    figsize=(10, 5)
)
plt.title('Ventas por mes por ao')
```

Noten como el eje Y empieza en 0 en el caso del grfico de barras

```python
import plotly.express as px
fig = px.line(
    por_mes,
    x=por_mes.index.values, # esto se hace de este modo ya que el indice es una columna sino hay q hacerlo reseteando el indice.
    y='sales'
)
fig.show()
```

```python
df['anio'] = df['date'].apply(lambda x: x[:4])
df
```

```python
# Filtro 2015 y las familias de productos que me interesan
df_2015 = df[
    (df['anio'] == '2015') &  
    (df['family'].isin(['SEAFOOD', 'MEATS', 'DAIRY']))
]
df_2015
```

```python
# Agrupo por mes y por familia
# las agrupaciones de multiples niveles debe ser en un formato de lista. []
agrupacion = df_2015[['fecha', 'family', 'sales']].groupby(['fecha', 'family']).sum()
agrupacion
```

```python
# Otra forma de hacerlo.

agrupacion2 = df_2015.groupby(['fecha','family']).agg(
    {
        'sales' : 'sum'
    }
)
agrupacion2
```

```python
# Pivoteo la tabla para que mis categorias pasen a ser mis columnas
pivote = agrupacion.unstack() 
# equivale a: agrupacion.reset_index(1).pivot(columns=['family'])
pivote
```

```python
#Otras opciones // index = 1 --> Familia
pivoteado = agrupacion.reset_index(1).pivot(columns='family')
pivoteado
```

```python
# Elimino un nivel en las columnas, no lo necesito  cabecera sales
pivote.columns = pivote.columns.droplevel()
pivote
```

```python
# Elimino un nivel en las columnas, no lo necesito  cabecera sales lo mismo q el paso anterior
pivoteado = agrupacion.reset_index(1).pivot(columns='family').droplevel(0,axis='columns')
pivoteado
```

```python
# Grafico de lineas de multiples categorias
pivote.plot.line(
    style='.-'
)
```

```python
# Grafico de lineas de multiples categorias
pivoteado.plot.line(
    style='.-'
)
```

```python
# Grafico de lineas de multiples categorias con Subplots
pivote.plot.line(
    style='.-',
    subplots=True
)
```

```python
# Grafico de lineas de multiples categorias con Subplots
pivoteado.plot.line(
    style='.-',
    subplots=True # subgraficos
)
```

# Grfico de Area
Combina la utilidad del grfico de lneas con la del grfico de barras. Me permite comparar diferencias entre magnitudes y hallar tendencias a lo largo de una progresin continua. Es por esto que en un grfico de Areas, el Eje Y si tiene que empezar en el "cero" (como el grfico de baras). Bsicamente es un grfico de lneas con el area bajo la curva de un color slido.   
**Eje X**: Variable categrica continua   
**Eje Y**: Variable numrica   
**Otros nombres**: Area Chart   
**Subtipos**: a) Stacked Area Chart (Grfico de Areas Apilados): Multiples lneas que acumulan sus valores numricos del Eje Y. b) Overlapped Area Chart (Grfico de Areas Agrupados): Multiples lneas que no acumulan sus valores numricos del Eje Y.



## Referencias
- https://datavizcatalogue.com/ES/metodos/grafica_de_area.html
- https://chartio.com/learn/charts/area-chart-complete-guide/

```python
# Grafico de Area Comun
por_mes.plot.area(
    y='sales'
)
```

```python
# Grafico de area apilado de multiples categorias
pivote.plot.area(
    style='.-',
    stacked=True
)
```

```python
# Grafico de area agrupado de multiples categorias
# El Eje Y no se acumula con cada lnea
pivote.plot.area(
    style='.-',
    stacked=False
)
```
