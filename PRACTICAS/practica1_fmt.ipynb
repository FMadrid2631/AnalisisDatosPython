```python
#Importo las librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#Busco el set de datos para trabajarlo
df = pd.read_csv('../data/comparasion_algoritmos.csv')
df.head()
```

```python
#Comienzo con la creacion del grafico
ax= df.plot(
    kind='bar',
    x='dataset',
    y=['algoritmo1','algoritmo2'],
    figsize=(12, 5)  
)

# aspecto del grafico
ax.grid(visible=True),
ax.set_xlabel("Dataset")
ax.set_ylabel("Rendimiento")
ax.set_title("Evaluacin de Algoritmo1 vs Algoritmo2")
ax.set_axisbelow(True) # ponemos por delante las barras sobre la grilla

#Legenda
ax.legend(labels=['Algoritmo 1', 'Algoritmo 2'], loc='upper left', bbox_to_anchor=(0, 1))
```
