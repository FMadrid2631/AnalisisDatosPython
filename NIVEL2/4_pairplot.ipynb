# Pair Plot
Es otro gráfico compuesto. Es un gráfico de "todos contra todos". Compara al mismo tiempo todas las variables numéricas de mi dataset a través de diversos Scatter plots arreglados en un grid (N x N). Siendo N el número de variables en mi dataset. En cada fila del grid, se compara en diversos scatter plot, una variable numérica VERSUS todas las demás. La diagonal del grid va a tener un histograma o un KDE de esa variable. (Ya que no tendría utilidad hacer un scatter plot donde ambos ejes correspondan a la misma variable)




### Cuando utilizar
- Quiero analizar si existe algun tipo de correlación entre alguna de todas mis variables numéricas. 
- Quiero observar la distribución de todas mis variables numéricas al mismo tiempo.
- Es uno de los gráficos más útiles al hacer análisis exploratorio de datos ya que me muestra un resúmen de todos mis datos numéricos.

### Consejos
- Si tengo muchas variables de interés el GRID se volverá demasiado grande. En estos casos puedo escoger un subconjunto de mis columnas para formar el Pair Plot.
- Puedo introducir una 3ra variable al gráfico con las siguientes técticas:
  - Diferentes colores o formas en los "puntos" para introducir una variable categórica.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Pair Plot con Seaborn
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automáticamente solo se utilizan las variables numéricas
    plot_kws={
        's': 10
    }
)
```

```python
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automáticamente solo se utilizan las variables numéricas
    hue='species',
    kind='kde' #hist
    #plot_kws={'s': 10 }
)
```

```python
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automáticamente solo se utilizan las variables numéricas
    hue='species',
    kind= 'hist'
    #plot_kws={'s': 10 }
)
```
