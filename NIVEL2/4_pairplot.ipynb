# Pair Plot
Es otro grfico compuesto. Es un grfico de "todos contra todos". Compara al mismo tiempo todas las variables numricas de mi dataset a travs de diversos Scatter plots arreglados en un grid (N x N). Siendo N el nmero de variables en mi dataset. En cada fila del grid, se compara en diversos scatter plot, una variable numrica VERSUS todas las dems. La diagonal del grid va a tener un histograma o un KDE de esa variable. (Ya que no tendra utilidad hacer un scatter plot donde ambos ejes correspondan a la misma variable)




### Cuando utilizar
- Quiero analizar si existe algun tipo de correlacin entre alguna de todas mis variables numricas. 
- Quiero observar la distribucin de todas mis variables numricas al mismo tiempo.
- Es uno de los grficos ms tiles al hacer anlisis exploratorio de datos ya que me muestra un resmen de todos mis datos numricos.

### Consejos
- Si tengo muchas variables de inters el GRID se volver demasiado grande. En estos casos puedo escoger un subconjunto de mis columnas para formar el Pair Plot.
- Puedo introducir una 3ra variable al grfico con las siguientes tcticas:
  - Diferentes colores o formas en los "puntos" para introducir una variable categrica.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Pair Plot con Seaborn
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automticamente solo se utilizan las variables numricas
    plot_kws={
        's': 10
    }
)
```

```python
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automticamente solo se utilizan las variables numricas
    hue='species',
    kind='kde' #hist
    #plot_kws={'s': 10 }
)
```

```python
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.pairplot(
    data=df, # Automticamente solo se utilizan las variables numricas
    hue='species',
    kind= 'hist'
    #plot_kws={'s': 10 }
)
```
