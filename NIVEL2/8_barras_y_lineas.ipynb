# Grfico de Barras con Lneas
Vamos a combinar un grfico de barras con un grfico de lneas.



### Cuando utilizar
- Quiero comparar cambios de magnitudes de una variable a travs del tiempo y aadir una segunda variable como un grfico de lneas que tenga una escala diferente


### Consejos
- Los mismos consejos que el grfico de barras y de lneas.
- Como consejo general, no recomiendo aadir mas de dos lneas al grfico. Se volvera muy complejo de leer.


## En esta clase
### Datos
Informacin sobre paises por ao: Expectativa de vida, Poblacin y GDP per Capita
### Objetivos
1. Comparar la expectativa de vida promedio en Africa a lo largo de los aos y al mismo tiempo ver como ha cambiado el GDP per Capita.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
#sns.set_palette("colorblind")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/paises_vida.csv', index_col='index')
df
```

```python
# Eliminamos filas NaNs, agrupamos por ao y por continente
df = df.dropna()
df_agrupado = df[df['continent'] == 'Africa'].groupby(['year']).agg({
    'lifeExp': np.mean,
    'pop': np.sum,
    'gdpPercap': np.mean
})
df_agrupado = df_agrupado.reset_index()
df_agrupado
```

```python
# Empecemos con el grafico de barras
sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey'
    #hue='continent'
)
```

```python
# Datos para el grafico de linea
# df_agrupado_pop = df[[]].groupby(['year']).agg({
#     'pop': np.sum,
# })
# df_agrupado_pop = df_agrupado_pop.reset_index()
# df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3))

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey',
    alpha=0.8,
    ax=ax1
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()

line_color = 'blue'
ax2.set_ylabel('Poblacin', color=line_color)
sns.lineplot(
    data=df_agrupado['gdpPercap'],
    alpha=1,
    marker='D',
    ax=ax2,
    linewidth=2,
    color=line_color
)

ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```

```python
df_agrupado = df.groupby(['year', 'continent']).agg({
    'lifeExp': np.mean,
    'pop': np.sum,
    'gdpPercap': np.sum
})
df_agrupado = df_agrupado.reset_index()
df_agrupado
```

```python
# Datos para el grafico de linea
df_agrupado_pop = df.groupby(['year']).agg({
    'pop': np.sum,
})
df_agrupado_pop = df_agrupado_pop.reset_index()
df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3))

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    alpha=0.8,
    hue='continent',
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()

line_color = 'brown'
ax2.set_ylabel('Poblacin', color=line_color)
sns.lineplot(
    data=df_agrupado_pop['pop'],
    alpha=1,
    marker='D',
    ax=ax2,
    linewidth=2,
    color=line_color
)

ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```
