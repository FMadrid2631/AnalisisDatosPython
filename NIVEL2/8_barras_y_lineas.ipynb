# Gráfico de Barras con Líneas
Vamos a combinar un gráfico de barras con un gráfico de līneas.



### Cuando utilizar
- Quiero comparar cambios de magnitudes de una variable a través del tiempo y añadir una segunda variable como un gráfico de líneas que tenga una escala diferente


### Consejos
- Los mismos consejos que el gráfico de barras y de līneas.
- Como consejo general, no recomiendo añadir mas de dos líneas al gráfico. Se volvería muy complejo de leer.


## En esta clase
### Datos
Información sobre paises por año: Expectativa de vida, Población y GDP per Capita
### Objetivos
1. Comparar la expectativa de vida promedio en Africa a lo largo de los años y al mismo tiempo ver como ha cambiado el GDP per Capita.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
#sns.set_palette("colorblind")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/paises_vida.csv', index_col='index')
df
```

```python
# Eliminamos filas NaNs, agrupamos por año y por continente
df = df.dropna()
df_agrupado = df[df['continent'] == 'Africa'].groupby(['year']).agg({
    'lifeExp': np.mean,
    'pop': np.sum,
    'gdpPercap': np.mean
})
df_agrupado = df_agrupado.reset_index()
df_agrupado
```

```python
# Empecemos con el grafico de barras
sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey'
    #hue='continent'
)
```

```python
# Datos para el grafico de linea
# df_agrupado_pop = df[[]].groupby(['year']).agg({
#     'pop': np.sum,
# })
# df_agrupado_pop = df_agrupado_pop.reset_index()
# df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3))

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey',
    alpha=0.8,
    ax=ax1
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()

line_color = 'blue'
ax2.set_ylabel('Población', color=line_color)
sns.lineplot(
    data=df_agrupado['gdpPercap'],
    alpha=1,
    marker='D',
    ax=ax2,
    linewidth=2,
    color=line_color
)

ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```

```python
df_agrupado = df.groupby(['year', 'continent']).agg({
    'lifeExp': np.mean,
    'pop': np.sum,
    'gdpPercap': np.sum
})
df_agrupado = df_agrupado.reset_index()
df_agrupado
```

```python
# Datos para el grafico de linea
df_agrupado_pop = df.groupby(['year']).agg({
    'pop': np.sum,
})
df_agrupado_pop = df_agrupado_pop.reset_index()
df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3))

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    alpha=0.8,
    hue='continent',
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()

line_color = 'brown'
ax2.set_ylabel('Población', color=line_color)
sns.lineplot(
    data=df_agrupado_pop['pop'],
    alpha=1,
    marker='D',
    ax=ax2,
    linewidth=2,
    color=line_color
)

ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```
